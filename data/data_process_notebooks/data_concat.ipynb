## 멀티 턴 

```python
df_1 = pd.read_csv('./multiturn_jieun.csv')
df_1['type'] = 'multi-2'
df_1
```

```python
df_2 = pd.read_csv('./jihwan_multi.csv')
df_2
```

```python
df_3 = pd.read_csv('./합성데이터_멀티턴_송인서.csv')
df_3['instruction'] = df_3['instruction'].apply(lambda x: x.replace("\\n", "\n"))
df_3['type'] = df_3['instruction'].apply(lambda x: 'multi-3' if x.count('질문:') == 3 else 'multi-2')
df_3
```

```python
df_4 = pd.read_csv('./multi_jh.csv')
df_4['type'] = 'multi-2'
df_4
```

```python
df_5 = pd.read_csv('./multi-turn_conversation_v1.csv')
df_5['instruction'] = df_5['instruction'].apply(lambda x: x.replace("\\n", "\n"))
df_5['type'] = df_5['instruction'].apply(lambda x: 'multi-3' if x.count('질문:') == 3 else 'multi-2')
df_5
```

```python
full_multi = pd.concat([df_1,df_2,df_3,df_4,df_5]).sample(frac=1, random_state=42).reset_index(drop=True)
full_multi
```

## 두 개 합치기

```python
full_df = pd.concat([full_single, full_multi]).sample(frac=1, random_state=42).reset_index(drop=True)
full_df['instruction'] = full_df['instruction'].apply(str)
full_df['output'] = full_df['output'].apply(str)
full_df['type'] = full_df['type'].apply(str)
```

```python
full_df.to_csv('./full_df.csv', index=False)
```

```python
import json

# 데이터프레임을 JSON Lines 형식으로 변환
json_data = full_df.to_dict(orient='records')

# ensure_ascii=False를 설정하여 유니코드를 이스케이프하지 않도록 함
json_str = "\n".join(json.dumps(item, ensure_ascii=False) for item in json_data)

with open('./Empathetic_data.jsonl', 'w', encoding='utf-8') as f:
    f.write(json_str)
```

## gpt-3.5-turbo multi-turn

```python
import pandas as pd


df = pd.read_csv('human_eval_multi_scores.csv')
df
```

```python
column_names = [
    '자연스러움 (1-5)',
    '흥미롭기 (1-5)',
    '공감 능력 (0-1)',
    '감정 파악 (1-5)',
    '반응 (1-5)',
    '사용자의 채팅 내용 사용 (0-1)',
    '맥락 유지 (1-5)',
    '대화 유도 (0-1)',
    '질문의 적절성 (1-5)',
    '전반적인 품질 (1-10)',
]

avg_scores = df[column_names].mean()
print(avg_scores)
new_row = pd.Series(["평균"] + list(avg_scores), index=df.columns)
df.loc[len(df)] = new_row
df.to_csv('human_eval_multi_scores.csv', index=False)
```

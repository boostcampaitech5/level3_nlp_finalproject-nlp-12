## 3턴 짜리 처리

```python
import pandas as pd

# 데이터를 저장할 리스트 초기화
data = []
# txt 파일 불러오기
with open('GPT4_synthetic_data_multi_turn_3.txt', 'r') as file:
    lines = file.readlines()

    # 대화 단위로 데이터 처리
    i = 0
    while i < len(lines):
        # 질문1~질문3 처리
        instruction = lines[i].strip() + "\n" + lines[i+1].strip() + "\n" + lines[i+3].strip() + "\n" + lines[i+4].strip() + "\n" + lines[i+6].strip()
        output = lines[i+7].strip().split('답변3')[1].lstrip(': ').rstrip('\n')
        data.append([instruction, output])
        i += 9


# 데이터프레임 생성
multi_3 = pd.DataFrame(data, columns=['instruction', 'output'])
multi_3['type'] = 'multi-3'
multi_3
```

## 5턴 짜리 처리

```python
import pandas as pd

# 데이터를 저장할 리스트 초기화
data = []
data_single = []
# txt 파일 불러오기
with open('GPT4_synthetic_data_multi_turn_5.txt', 'r') as file:
    lines = file.readlines()

    # 대화 단위로 데이터 처리
    i = 0
    while i < len(lines):
        # 질문1~질문3 처리
        if lines[i].startswith('질문1'):
            instruction = lines[i].strip() + "\n" + lines[i+1].strip() + "\n" + lines[i+3].strip() + "\n" + lines[i+4].strip() + "\n" + lines[i+6].strip()
            output = lines[i+7].strip().split('답변3')[1].lstrip(': ').rstrip('\n')
            data.append([instruction, output])
            i += 8
        # 질문4~질문5 처리
        elif lines[i].startswith('질문4'):
            if lines[i+3].startswith('질문5'):
                instruction = lines[i].strip() + "\n" + lines[i+1].strip() + "\n" + lines[i+3].strip()
                output = lines[i+4].split('답변5')[1].lstrip(': ').rstrip('\n')
                data.append([instruction, output])
                i += 5
            # 생성 중 잘려서 질문4-답변4까지만 생성된 경우(질문5-답변5가 없음) 4를 싱글턴으로 저장
            else:
                instruction = lines[i].split('질문4')[1].lstrip(': ').rstrip('\n')
                output = lines[i+1].split('답변4')[1].lstrip(': ').rstrip('\n')
                data_single.append([instruction, output])
                i += 3
        else:
            i += 1

# 데이터프레임 생성
df = pd.DataFrame(data, columns=['instruction', 'output'])
df_single = pd.DataFrame(data_single, columns=['instruction', 'output'])

# 저장
df.to_csv('./jihwan_multi.csv', index=False)
df
```

# 2. 싱글 턴 대화

```python
def txt_to_df(text):
    # "질문:"과 "답변:"로 시작하는 줄을 기준으로 분할
    # 질문 답변 콜론형태가 이상할 경우 수정
    text = re.sub('질문\s*:', '질문:', text)
    text = re.sub('답변\s*:', '답변:', text)
    text = re.sub('질문\d*\s*:', '질문:', text)
    text = re.sub('답변\d*\s*:', '답변:', text)
    split_data = re.split("질문: |답변: ", text)

    # 첫 번째 원소가 빈 문자열이므로 제거
    split_data = split_data[1:]

    # '질문'과 '답변' 리스트 생성
    questions = [split_data[i].strip() for i in range(0, len(split_data), 2)]
    answers = [split_data[i].strip() for i in range(1, len(split_data), 2)]

    # 데이터프레임 생성
    df = pd.DataFrame({
        'instruction' : questions,
        'output' : answers
    })

    return df
```

```python
with open('./GPT4_synthetic_data.txt', 'r') as f:
    data = f.read()

full_df = txt_to_df(data)
single_turn = pd.concat([full_df, df_single]).reset_index(drop=True)
```

```python
single_turn.to_csv('./jihwan_single.csv', index=False)
```

```python
full_df = txt_to_df(data)
print(full_df)

# csv로 저장
full_df.to_csv('./jihwan_data.csv', index=False)

import json

# 데이터프레임을 JSON Lines 형식으로 변환
json_data = full_df.to_dict(orient='records')

# ensure_ascii=False를 설정하여 유니코드를 이스케이프하지 않도록 함
json_str = "\n".join(json.dumps(item, ensure_ascii=False) for item in json_data)

# 파일에 쓰기
with open('./Empathetic_data.jsonl', 'w', encoding='utf-8') as f:
    f.write(json_str)
```
